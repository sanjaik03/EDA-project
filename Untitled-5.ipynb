```python
x=[1,2,3,4,5,6,7,8,9]
y=[11,22,33,44,55,66,77,88,99]
plt.plot(x,y)
plt.title("zzz")
plt.xlabel("y")
plt.ylabel("a")
plt.show()
```

```python
x1 = [23,72,98,88,78,68,44,54,81]
x2 = [47,58,72,99,100,101,112,24,25]

y1 = [66,54,78,62,13,11,1,2,3]
y2 = [72,98,93,94,67,54,32,33,50]

plt.plot(x1,y1,'r')
plt.plot(x2,y2,'g')
plt.plot(x1,x2,'ro')
plt.plot(y1,y2,'go')
plt.title('zzz')
plt.xlabel('yyy')
plt.ylabel('xxx')
plt.show()
```

```python
x1 = [23,72,98,88,78,68,44,54,81]
x2 = [47,58,72,99,100,101,112,24,25]

y1 = [66,54,78,62,13,11,1,2,3]
y2 = [72,98,93,94,67,54,32,33,50]

plt.bar(x1,y1,color='b',label = 'profit')
plt.bar(x2,y2,color = 'g',label = 'tax')
plt.legend()

plt.title('zzz')
plt.xlabel('yyy')
plt.ylabel('xxx')
plt.show()
```

```python
x= [7,8,5,2,4,3,1,8,12]
y =[12,11,5,8,9,12,14,2,1]

plt.scatter(x,y,marker="*",color="b")
plt.title('XXX')
plt.xlabel('age')
plt.ylabel('Range')
plt.grid()
plt.show()
```

```python
x = [52, 60, 80, 88, 76, 90]
y = ['eng', 'tamil', 'maths', 'social', 'phy', 'che']
c = ['r', 'g', 'c', 'b', 'm', 'y'] 

plt.pie(x, labels=y, colors=c, autopct='%1.1f%%')

plt.title('xxx')
plt.show()
```
