```python
#BOX plot
data=df[df['price']<1500]
sns.boxplot(x='price',data=df)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="price", bins=100)
plt.title("Price Distribution")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.histplot(data=df, x="availability_365", bins=100)
plt.title("availability_365")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
df['price per bed'] = df['price']/df['beds']
df.head()
```

```python
#Bar plot
sns.barplot(data=df,x="neighbourhood_group",y='price',hue='room_type')
plt.show()
```

```python
#Scatter plot
plt.figure(figsize=(8,5))
plt.title('scatter plot')
plt.xlabel('x')
plt.ylabel('y')

sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.show()
```

```python
#Pair plot
sns.pairplot(data=df,vars=['price', 'minimum_nights', 'number_of_reviews'],hue='room_type')
```

```python
#HeatMap
cr=df[['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
cr
plt.figure(figsize=(8,6))
sns.heatmap(data=cr,annot=True)
```
