```python
data=[11,22,33,44,55,66,77,88,99]
plt.hist(data,bins=5,color="tomato",edgecolor='black')
plt.title("histogram")
plt.xlabel("xx")
plt.ylabel("yy")
plt.show()
```

```python
a=[12,23,34,45,56]
b=[67,78,89,91,10]
plt.plot(a,label="Product A",marker="o")
plt.plot(b,label="Product B",marker="s")
plt.title("Total number of days")
plt.show()
```

```python
size=[5,8,2,6,4]
labels=['east','west','north','south','central']
plt.pie(size,labels=labels,autopct="%1.1f%%",startangle=90)
plt.title("pole")
```

```python
a=[12,23,34,45,56]
b=[67,78,89,91,10]

plt.hexbin(a,b,gridsize=30,cmap="Blues")
plt.colorbar(label='count')
plt.title('xx')
```

```python
import matplotlib.pyplot as plt

months = ['Mar', 'Jun', 'Jul', 'Aug', 'Sep']   
sales = [100, 150, 200, 250, 300]  

plt.fill_between(months, sales, color='pink', alpha=0.79)
plt.plot(months, sales, color='blue')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```
