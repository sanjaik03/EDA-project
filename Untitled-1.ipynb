```python
import seaborn as sns
import matplotlib as plt
import pandas as pd
```

```python
data = {
    "employee": ['bbb','aaa','zzz','xxx','yyy','ccc','ddd','qqq','rrr'],
    "department": ['it','sales','hr','hr','it','hr','it','sales','it'],
    "salary": [15000, 10000, 11000, 11000, 15000, 11000, 15000, 10000, 14000],
    "exp": [3, 1, 2, 2, 4, 1, 5, 1, 3],          
    "projects": [4, 2, 3, 3, 5, 2, 6, 1, 4],     
    "performance": [78, 55, 68, 70, 90, 60, 95, 45, 80]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
sns.scatterplot(x='exp',y='salary',data=df)
```

```python
sns.lineplot(x='exp',y='performance',markers='s',data=df)
```

```python
sns.barplot(x='exp',y='salary',data=df)
```

```python
sns.countplot(x='exp',data=df)
```

```python
sns.boxplot(x='salary',y='performance',data=df)
```

```python
sns.violinplot(x='salary',y='exp',data=df)
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
sns.pairplot(df,hue='department')
```

```python
sns.histplot(df['salary'],bins=10,kde=True)
```
